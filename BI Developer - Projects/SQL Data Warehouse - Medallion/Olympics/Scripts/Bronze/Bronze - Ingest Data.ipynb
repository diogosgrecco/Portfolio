```python
%%pyspark


dfa = spark.read.option("header", "true").csv("Files/Raw Data/Athletes.csv")
# remove unwanted spaces from column names
dfa = dfa.toDF(*[col.replace(" ", "") for col in dfa.columns])

dfc = spark.read.option("header", "true").csv("Files/Raw Data/Coaches.csv")
# remove unwanted spaces from column names
dfc = dfc.toDF(*[col.replace(" ", "") for col in dfc.columns])

dfe = spark.read.option("header", "true").csv("Files/Raw Data/EntriesGender.csv")
# remove unwanted spaces from column names
dfe = dfe.toDF(*[col.replace(" ", "") for col in dfe.columns])

dfm = spark.read.option("header", "true").csv("Files/Raw Data/Medals.csv")
# remove unwanted spaces from column names
dfm = dfm.toDF(*[col.replace(" ", "") for col in dfm.columns])

dft = spark.read.option("header", "true").csv("Files/Raw Data/Teams.csv")
# remove unwanted spaces from column names
dft = dft.toDF(*[col.replace(" ", "") for col in dft.columns])


# create temp table in order to be able to use them in sql
dfa.write.mode("overwrite").saveAsTable("Bronze.temp_athletes")
dfc.write.mode("overwrite").saveAsTable("Bronze.temp_coaches")
dfe.write.mode("overwrite").saveAsTable("Bronze.temp_entriesGender")
dfm.write.mode("overwrite").saveAsTable("Bronze.temp_medals")
dft.write.mode("overwrite").saveAsTable("Bronze.temp_teams");
```

```python
%%sql

-------------------------------------------------
-- Create Table AND Insert Data As Is From csv --
-------------------------------------------------

DROP TABLE IF EXISTS Bronze.athletes;

CREATE TABLE Bronze.athletes 
AS
SELECT *
FROM Bronze.temp_athletes;


DROP TABLE IF EXISTS Bronze.coaches;

CREATE TABLE Bronze.coaches 
AS
SELECT *
FROM Bronze.temp_coaches;


DROP TABLE IF EXISTS Bronze.entriesGender;

CREATE TABLE Bronze.entriesGender 
AS
SELECT *
FROM Bronze.temp_entriesGender;


DROP TABLE IF EXISTS Bronze.medals;

CREATE TABLE Bronze.medals 
AS
SELECT *
FROM Bronze.temp_medals;


DROP TABLE IF EXISTS Bronze.teams;

CREATE TABLE Bronze.teams 
AS
SELECT *
FROM Bronze.temp_teams;

```
