```sql
TRUNCATE TABLE Silver.athletes;

INSERT INTO Silver.athletes
SELECT row_number() OVER(ORDER BY PersonName) AS athlete_id, TRIM(PersonName) AS athlete_name, UPPER(TRIM(Country)) AS country, UPPER(TRIM(Discipline)) AS discipline
FROM Bronze.athletes;


TRUNCATE TABLE Silver.coaches;

INSERT INTO Silver.coaches
SELECT row_number() OVER(ORDER BY Name) AS coach_id, TRIM(Name) AS coach_name, UPPER(TRIM(Country)) AS country, UPPER(TRIM(Discipline)) AS discipline, UPPER(TRIM(Event)) as event
FROM Bronze.coaches;


TRUNCATE TABLE Silver.teams;

INSERT INTO Silver.teams
SELECT row_number() OVER(ORDER BY TeamName, Discipline, Country, Event) AS team_id, TRIM(TeamName) AS team_name, UPPER(TRIM(Discipline)) AS discipline, UPPER(TRIM(Country)) AS country, UPPER(TRIM(Event)) AS event
FROM Bronze.teams;


TRUNCATE TABLE Silver.entries_gender;

INSERT INTO Silver.entries_gender
SELECT row_number() OVER(ORDER BY Discipline) AS entry_id, UPPER(TRIM(Discipline)) AS discipline, CAST(Female AS INT) AS female, CAST(Male AS INT) AS male, CAST(Total AS INT) AS total
FROM Bronze.entriesgender;


TRUNCATE TABLE Silver.medals;

INSERT INTO Silver.medals
SELECT row_number() OVER(ORDER BY CAST(Rank AS INT)) AS medal_id, CAST(Rank AS INT) AS rank, UPPER(TRIM(TeamCountry)) AS team_country, CAST(Gold AS INT) AS gold_medals, CAST(Silver AS INT) AS silver_medals, CAST(Bronze AS INT) AS bronze_medals, CAST(Total AS int) AS total_medals, CAST(RankbyTotal AS INT) AS rank_by_total
FROM Bronze.medals;
```
