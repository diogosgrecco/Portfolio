```sql
----------------------------------
-- Create Gold Table Structures --
----------------------------------

-- DIM Tables --
DROP TABLE IF EXISTS Gold.dim_countries;

CREATE TABLE Gold.dim_countries (
    country_id INT,
    country_name STRING
);


DROP TABLE IF EXISTS Gold.dim_disciplines;

CREATE TABLE Gold.dim_disciplines (
    discipline_id INT,
    discipline_name STRING
);


DROP TABLE IF EXISTS Gold.dim_events;

CREATE TABLE Gold.dim_events (
    event_id INT,
    event_name STRING
);


-- FACT Tables --
DROP TABLE IF EXISTS Gold.fact_medals_by_country;

CREATE TABLE Gold.fact_medals_by_country (
    country_id INT,
    gold_count INT,
    silver_count INT,
    bronze_count INT,
    total_medals INT,
    rank INT,
    rank_by_total INT
);


DROP TABLE IF EXISTS Gold.fact_medals_by_country;

CREATE TABLE Gold.fact_medals_by_country (
    discipline_id INT,
    total_female INT,
    total_male INT,
    total INT
);


DROP TABLE IF EXISTS Gold.fact_medals_by_country;

CREATE TABLE Gold.fact_medals_by_country (
    athlete_id INT,
    country_id INT,
    discipline_id INT
);
```
